```python
df0 = recom.recomendation('dynamics of magmatic intrusions').head(25)
links = df0.link.tolist()
qry = 'select time,place,linkp ' +\
            'from papers ' +\
            'where linkp in (%s)' % (', '.join(['?'] * len(links)))
res = recom.query_db(qry, links)
df = pd.DataFrame([[row[f] for f in res[0].keys()]
                           for row in res], columns=res[0].keys())
df = df.rename(columns={'linkp': 'link'})
df = pd.merge(df, df0[['link', 'score','title']],
                      on='link', how='outer')
df.score = map(lambda x: float(x), df.score)
df = df.sort_values(by='score',ascending=False)
df['room'] = map(get_room,df.place)
df['type'] = map(get_type_pres,df.place)
```

```python
def get_type_pres(x):
    room = x.split('-')[-1].strip()
    if room =="Poster Hall":
        return 'poster'
    else:
        return 'oral'
def get_room(x):
    room = x.split('-')[-1].strip()
    if room == 'Poster Hall':
        return room
    else:
        return 'Room %s'%(room)
```
