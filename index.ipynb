# Abstract-based recommendation for the GeoScience community

American Geophysical  Union (AGU)  meeting is a  geoscience conference
hold each year around Christmas in  San Francisco.  With nearly 24 000
attendees, AGU  Fall Meeting  is the largest  Earth and  Space science
meeting in  the world.  As  such, it  represents an appealing  tool to
explore the important trends in the geoscience academic world.

I have been  there every year since  the beginning of my  PhD and each
year, I ends up with the  same conclusion; lost in translation between
24 000 contributions, it  is very hard to focus on who  to talk to and
what presentations to  see given a specific subject.   That's why, last
December,  I  decided to  develop  a  tool  which  would aim  to  both
facilitate   the  navigation   within  the   conference  program   and
collaboration  between  researchers  based   on  the  content  of  the
conference.  Especially  for newcomers  in the  field like  myself, it
would be  very nice if such  tool existed, simply to  identify who and
where people are working on a specific thematic.
